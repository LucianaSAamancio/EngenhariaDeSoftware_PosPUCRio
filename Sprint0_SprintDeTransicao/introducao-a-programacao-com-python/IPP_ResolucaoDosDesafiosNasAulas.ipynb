Desafio: o que é impresso por esta função?

```python
s = input('Digite uma string: ')
def desafio(s):
  for c in reversed(s):
      print(c)
  return
```

A função desafio(s) recebe uma string s como argumento e imprime cada caractere da string em ordem reversa.

O loop for c in reversed(s) percorre os caracteres da string começando do último até o primeiro, e a função print(c) exibe cada caractere na tela. No final, a função não retorna nenhum valor, pois o return está vazio.

Por exemplo, se você chamar

desafio("Olá")

a saída será:

á

l

O

**Solução do Desafio**

```python
# Função
def desafio(s):
  for c in reversed(s):
      print(c)
  return

# Bloco principal
s = input('Digite uma string: ')
desafio(s)
```

E se, em vez de reversed(s), for digitado s[::-1]? O resultado é exatamente igual.

```python
# Função
def desafio(s):
  for c in s[::-1]:
      print(c)
  return

# Bloco principal
s = input('Digite uma string: ')
desafio(s)
```
