Desafio: o que  impresso por esta funo?

```python
s = input('Digite uma string: ')
def desafio(s):
  for c in reversed(s):
      print(c)
  return
```

A funo desafio(s) recebe uma string s como argumento e imprime cada caractere da string em ordem reversa.

O loop for c in reversed(s) percorre os caracteres da string comeando do ltimo at o primeiro, e a funo print(c) exibe cada caractere na tela. No final, a funo no retorna nenhum valor, pois o return est vazio.

Por exemplo, se voc chamar

desafio("Ol")

a sada ser:



l

O

**Soluo do Desafio**

```python
# Funo
def desafio(s):
  for c in reversed(s):
      print(c)
  return

# Bloco principal
s = input('Digite uma string: ')
desafio(s)
```

E se, em vez de reversed(s), for digitado s[::-1]? O resultado  exatamente igual.

```python
# Funo
def desafio(s):
  for c in s[::-1]:
      print(c)
  return

# Bloco principal
s = input('Digite uma string: ')
desafio(s)
```
