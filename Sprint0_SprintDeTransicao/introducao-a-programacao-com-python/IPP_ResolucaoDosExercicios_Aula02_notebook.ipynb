1 . Crie a função somaAlgarismos, que receba como parâmetro um número

Crie a função criaSenha, que receba três números inteiros como parâmetros representando uma data de nascimento (dia, mes e ano) e retorne a string resultante da concatenação da soma dos algarismos do dia com a soma dos algarismos do mês com a soma dos dois últimos algarismos do ano; para calcular cada soma, utilize a função somaAlgarismos; se, por exemplo, a função receber os números 17, 12 e 1921, deverá retornar a string 833.

Teste a sua função para os valores 6, 9 e 10.

```python
# Função
def somaImpares(n):
    soma = 0
    cont = 0
    impar = 1
    while cont < n:
        soma += impar
        impar += 2
        cont += 1
    return soma

# Bloco principal - testes da função
print(f'Soma dos 6 primeiros ímpares: {somaImpares(6)}')
print(f'Soma dos 9 primeiros ímpares: {somaImpares(9)}')
print(f'Soma dos 10 primeiros ímpares: {somaImpares(10)}')
```

**DESAFIO 7**

Escreva uma função chamada exibe2Maiores que receba dois valores inteiros como parâmetros representando o início e o fim de um intervalo ([início, fim]), gere aleatoriamente 25 inteiros nesse intervalo e exiba os dois maiores inteiros gerados em ordem decrescente.

Teste a sua função para os intervalos [30, 90] e [110, 8752]. Execute o seu programa várias vezes e observe os resultados.

```python
# Módulo
from random import randint

# Função
def exibe2Maiores(inicio, fim):
    maior1 = maior2 = inicio - 1    # equivalente a: maior1 = inicio - 1
                                    #                maior2 = inicio - 1
    cont = 0
    while cont < 25:
        nAleatorio = randint(inicio, fim)
        if nAleatorio > maior1:
            maior2 = maior1
            maior1 = nAleatorio
        elif nAleatorio > maior2:
            maior2 = nAleatorio
        cont += 1
    print(f'Dois maiores valores em ordem decrescente: {maior1}, {maior2}')
    return

# Bloco principal - testes da função
print('Intervalo [30, 90]')
exibe2Maiores(30, 90)
print('\nIntervalo [110, 8752]')
exibe2Maiores(110, 8752)
```

Altere o programa do desafio anterior para que, após processar todos os alunos da turma, também exiba a média aritmética dos alunos que obtiveram média menor que 3.0 na disciplina.

```python
# Funções
def calculaMedia(n1, n2, peso1, peso2):
    return (n1 * peso1 + n2 * peso2) / (peso1 + peso2)

def exibeMedias(mediaTestes, mediaProvas, mediaDisciplina):
    print(f'\nMédia dos testes: {mediaTestes:.1f}')
    print(f'Média das provas: {mediaProvas:.1f}')
    print(f'Média da disciplina: {mediaDisciplina:.1f}')
    print('\nSituação final do aluno: ', end='')
    if mediaDisciplina >= 7.0:
        print('Aprovado\n')
    elif mediaDisciplina >= 3.0:
        print('Em prova final\n')
    else:
        print('Reprovado\n1')
    return

# Bloco principal
peso1 = float(input('Digite a nota do peso 1: '))
peso2 = float(input('Digite a nota do peso 2: '))
totMenorQ3 = somaMenorQ3 = 0
contAlunos = 1
while contAlunos <= 7:
    print(f'Notas do {contAlunos}º aluno')
    teste1 = float(input('Digite a nota do teste 1: '))
    teste2 = float(input('Digite a nota do teste 2: '))
    prova1 = float(input('Digite a nota do prova 1: '))
    prova2 = float(input('Digite a nota do prova 2: '))

    # Cálculo das médias
    mediaTestes = calculaMedia(teste1, teste2, peso1, peso2)
    mediaProvas = calculaMedia(prova1, prova2, peso1, peso2)
    mediaDisciplina = calculaMedia(mediaTestes, mediaProvas, peso1, peso2)

    # Exibição do resultado
    exibeMedias(mediaTestes, mediaProvas, mediaDisciplina)

    # Totalizando alunos com media na disciplina < 3.0
    if mediaDisciplina < 3.0:
        somaMenorQ3 += mediaDisciplina
        totMenorQ3 += 1
    contAlunos += 1

# Exibição do resultado final
if totMenorQ3 == 0:
    print('\n\nNenhum aluno obteve média menor que 3.0 na disciplina.')
else:
    print(f'\n\nMédia dos alunos que obtiveram média na disciplina ', end ='')
    print(f'menor que 3: {somaMenorQ3 / totMenorQ3:.1f}')
```
