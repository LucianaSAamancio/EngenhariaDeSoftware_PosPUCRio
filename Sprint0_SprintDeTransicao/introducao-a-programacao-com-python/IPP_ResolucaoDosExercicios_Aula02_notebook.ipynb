```python
# Funo
def somaImpares(n):
    soma = 0
    cont = 0
    impar = 1
    while cont < n:
        soma += impar
        impar += 2
        cont += 1
    return soma

# Bloco principal - testes da funo
print(f'Soma dos 6 primeiros mpares: {somaImpares(6)}')
print(f'Soma dos 9 primeiros mpares: {somaImpares(9)}')
print(f'Soma dos 10 primeiros mpares: {somaImpares(10)}')
```

Teste a sua funo para os intervalos [30, 90] e [110, 8752]. Execute o seu programa vrias vezes e observe os resultados.

```python
# Mdulo
from random import randint

# Funo
def exibe2Maiores(inicio, fim):
    maior1 = maior2 = inicio - 1    # equivalente a: maior1 = inicio - 1
                                    #                maior2 = inicio - 1
    cont = 0
    while cont < 25:
        nAleatorio = randint(inicio, fim)
        if nAleatorio > maior1:
            maior2 = maior1
            maior1 = nAleatorio
        elif nAleatorio > maior2:
            maior2 = nAleatorio
        cont += 1
    print(f'Dois maiores valores em ordem decrescente: {maior1}, {maior2}')
    return

# Bloco principal - testes da funo
print('Intervalo [30, 90]')
exibe2Maiores(30, 90)
print('\nIntervalo [110, 8752]')
exibe2Maiores(110, 8752)
```

```python
# Funes
def calculaMedia(n1, n2, peso1, peso2):
    return (n1 * peso1 + n2 * peso2) / (peso1 + peso2)

def exibeMedias(mediaTestes, mediaProvas, mediaDisciplina):
    print(f'\nMdia dos testes: {mediaTestes:.1f}')
    print(f'Mdia das provas: {mediaProvas:.1f}')
    print(f'Mdia da disciplina: {mediaDisciplina:.1f}')
    print('\nSituao final do aluno: ', end='')
    if mediaDisciplina >= 7.0:
        print('Aprovado\n')
    elif mediaDisciplina >= 3.0:
        print('Em prova final\n')
    else:
        print('Reprovado\n1')
    return

# Bloco principal
peso1 = float(input('Digite a nota do peso 1: '))
peso2 = float(input('Digite a nota do peso 2: '))
totMenorQ3 = somaMenorQ3 = 0
contAlunos = 1
while contAlunos <= 7:
    print(f'Notas do {contAlunos} aluno')
    teste1 = float(input('Digite a nota do teste 1: '))
    teste2 = float(input('Digite a nota do teste 2: '))
    prova1 = float(input('Digite a nota do prova 1: '))
    prova2 = float(input('Digite a nota do prova 2: '))

    # Clculo das mdias
    mediaTestes = calculaMedia(teste1, teste2, peso1, peso2)
    mediaProvas = calculaMedia(prova1, prova2, peso1, peso2)
    mediaDisciplina = calculaMedia(mediaTestes, mediaProvas, peso1, peso2)

    # Exibio do resultado
    exibeMedias(mediaTestes, mediaProvas, mediaDisciplina)

    # Totalizando alunos com media na disciplina < 3.0
    if mediaDisciplina < 3.0:
        somaMenorQ3 += mediaDisciplina
        totMenorQ3 += 1
    contAlunos += 1

# Exibio do resultado final
if totMenorQ3 == 0:
    print('\n\nNenhum aluno obteve mdia menor que 3.0 na disciplina.')
else:
    print(f'\n\nMdia dos alunos que obtiveram mdia na disciplina ', end ='')
    print(f'menor que 3: {somaMenorQ3 / totMenorQ3:.1f}')
```
